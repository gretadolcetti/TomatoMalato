```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_pie_chart(subfolder_path):
    # Inizializziamo due liste: labels conterrà i nomi delle classi, sizes conterrà il numero di elementi di ogni classe
    labels = []
    sizes = []

    # Iteriamo sulla sottocartella, salvando i nomi delle sottocartelle (che rappresentano le classi) e il numero di elementi per ogni classe
    for subfolder in os.listdir(subfolder_path):
        subfolder_name = subfolder
        subfolder_size = len(os.listdir(os.path.join(subfolder_path, subfolder)))
        labels.append(subfolder_name)
        sizes.append(subfolder_size)

    # Calcoliamo la percentuale di ogni classe rispetto al totale
    sizes = np.array(sizes)
    percentages = (sizes / sizes.sum()) * 100

    # Definiamo 11 colori tutti diversi tra loro e ben visibili
    colors=['#f44336', '#e91e63', '#9c27b0', '#673ab7', '#3f51b5', '#2196f3', '#03a9f4', '#00bcd4', '#009688', '#4caf50', '#8bc34a']

    # Definiamo la figura e il grafico a torta
    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False, startangle=90, colors=colors)
    ax.set_title(subfolder_path + f": totale immagini {sizes.sum()}")
    ax.axis('equal')

    # Stampiamo le statistiche 
    for i, label in enumerate(labels):
        print(f'Classe {label}: {sizes[i]} elementi ({percentages[i]:.1f}%)')
    print(f'Totale: {sizes.sum()} elementi')
```

```python
# Percorso della cartella principale
main_folder = 'Dataset'

# Percorso delle sottocartelle
subfolders = ['test', 'train', 'valid']

# Iteriamo su ogni sottocartella
for subfolder in subfolders:
    subfolder_path = os.path.join(main_folder, subfolder)
    print(f'Analisi della sottocartella {subfolder}:')
    plot_pie_chart(subfolder_path)
```
